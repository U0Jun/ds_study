```python
from bs4 import BeautifulSoup
import requests
import lxml
```

```python
def contents_url(url) :
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'lxml')
    metadata = soup.find_all('div', class_ = 'ranking_headline', limit =3)
    
    for i in metadata :
        print("http://news.naver.com" + i.a.get('href'))
```
